```python
# Step 1: import libraries
%pip install --upgrade openpyxl
%pip install --upgrade pandas
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
import matplotlib.pyplot as plt
import numpy as np

# Step 2: load and clean the Data
data = pd.read_excel('final_merged_with_NULLs.xlsx')
data.replace("NULL", np.nan, inplace=True)

# Step 3: Aggregate Indicators by Country
aggregated_data = data.groupby('Country').agg({
    'Human Development Index (value)': 'mean',
    'Expected Years of Schooling (years)': 'mean',
    'Mean Years of Schooling (years)': 'mean',
    'Gross National Income Per Capita (2017 PPP$)': 'mean',
    'Life Expectancy at Birth (years)': 'mean'
}).reset_index()

# Step 4: Select features for clustering
features = [
    'Human Development Index (value)', 'Expected Years of Schooling (years)', 
    'Mean Years of Schooling (years)', 'Gross National Income Per Capita (2017 PPP$)', 
    'Life Expectancy at Birth (years)'
]
X = aggregated_data[features].fillna(aggregated_data[features].median())

# Step 5: Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 6: Determine Optimal Number of Clusters
sse = []
silhouette_scores = []

# Test different numbers of clusters
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

# Plot the Elbow Method for SSE
plt.figure(figsize=(10, 5))
plt.plot(range(2, 11), sse, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
plt.title('Elbow Method for Optimal k')
plt.show()

# Plot the Silhouette Scores
plt.figure(figsize=(10, 5))
plt.plot(range(2, 11), silhouette_scores, marker='o', color='orange')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal k')
plt.show()

# Step 7: Apply KMeans with Optimal k (e.g., k=3 based on plots)
optimal_k = 6
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
aggregated_data['Cluster'] = kmeans.fit_predict(X_scaled)

# Step 8: Analyze and Interpret Clusters
cluster_summary = aggregated_data.groupby('Cluster')[features].mean()
print("Cluster Centers (Averaged Features by Cluster):\n", cluster_summary)

# Display countries in each cluster
for cluster_num in sorted(aggregated_data['Cluster'].unique()):
    print(f"\nCountries in Cluster {cluster_num}:")
    countries_in_cluster = aggregated_data[aggregated_data['Cluster'] == cluster_num]['Country'].tolist()
    print(countries_in_cluster)

# Save clustered data to a new file
new_file_path = 'clustering_by_country.csv'
aggregated_data.to_csv(new_file_path, index=False)

# Step 9: Evaluate Clustering Model with Additional Metrics
dbi = davies_bouldin_score(X_scaled, aggregated_data['Cluster'])
chi = calinski_harabasz_score(X_scaled, aggregated_data['Cluster'])
sil_score = silhouette_score(X_scaled, aggregated_data['Cluster'])

print("Davies-Bouldin Index (DBI):", dbi)
print("Calinski-Harabasz Index (CHI):", chi)
print("Silhouette Score:", sil_score)
```

```python
# Step 1: Import Libraries
%pip install --upgrade openpyxl
%pip install --upgrade pandas
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
import matplotlib.pyplot as plt
import numpy as np

# Step 2: Load and Clean the Data
data = pd.read_excel('final_merged_with_NULLs.xlsx')
data.replace("NULL", np.nan, inplace=True)  # Replace "NULL" with NaN

# Step 3: Aggregate Indicators by IncomeGroup
aggregated_data = data.groupby('IncomeGroup').agg({
    'Human Development Index (value)': 'mean',
    'Expected Years of Schooling (years)': 'mean',
    'Mean Years of Schooling (years)': 'mean',
    'Gross National Income Per Capita (2017 PPP$)': 'mean',
    'Life Expectancy at Birth (years)': 'mean'
}).reset_index()

# Step 4: Select Features for Clustering
features = [
    'Human Development Index (value)', 'Expected Years of Schooling (years)', 
    'Mean Years of Schooling (years)', 'Gross National Income Per Capita (2017 PPP$)', 
    'Life Expectancy at Birth (years)'
]
X = aggregated_data[features].fillna(aggregated_data[features].median())

# Step 5: Standardize the Data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 6: Determine Optimal Number of Clusters
sse = []
silhouette_scores = []
cluster_range = range(2, min(5, len(aggregated_data)) + 1)

# Test different numbers of clusters within the valid range
for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)
    
    if k < len(aggregated_data):
        silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

# Plot the Elbow Method for SSE
plt.figure(figsize=(10, 5))
plt.plot(cluster_range, sse, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
plt.title('Elbow Method for Optimal k')
plt.show()

# Plot the Silhouette Scores if valid
if silhouette_scores:
    plt.figure(figsize=(10, 5))
    plt.plot(range(2, len(silhouette_scores) + 2), silhouette_scores, marker='o', color='orange')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Silhouette Score')
    plt.title('Silhouette Score for Optimal k')
    plt.show()

# Step 7: Apply KMeans with Optimal k (e.g., k=3 based on plots)
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
aggregated_data['Cluster'] = kmeans.fit_predict(X_scaled)

# Step 8: Analyze and Interpret Clusters
cluster_summary = aggregated_data.groupby('Cluster')[features].mean()
print("Cluster Centers (Averaged Features by Cluster):\n", cluster_summary)

# Display income groups in each cluster
for cluster_num in sorted(aggregated_data['Cluster'].unique()):
    print(f"\nIncome Groups in Cluster {cluster_num}:")
    income_groups_in_cluster = aggregated_data[aggregated_data['Cluster'] == cluster_num]['IncomeGroup'].tolist()
    print(income_groups_in_cluster)

# Save clustered data to a new file
new_file_path = 'clustering_by_income_group.csv'
aggregated_data.to_csv(new_file_path, index=False)

# Step 9: Evaluate Clustering Model with Additional Metrics
dbi = davies_bouldin_score(X_scaled, aggregated_data['Cluster'])
chi = calinski_harabasz_score(X_scaled, aggregated_data['Cluster'])
sil_score = silhouette_score(X_scaled, aggregated_data['Cluster']) if optimal_k < len(aggregated_data) else "N/A"

print("Davies-Bouldin Index (DBI):", dbi)
print("Calinski-Harabasz Index (CHI):", chi)
print("Silhouette Score:", sil_score)
```
