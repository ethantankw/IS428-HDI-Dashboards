purpose of this code is to group countries with similar socio-economic characteristics (e.g., schooling, income, life expectancy). these clusters can help identify shared needs or strategies within the groups, allowing for a more targeted improvement strategies based on the unique characteristics of each of the clusters.

```python
# Step 1: import libraries
%pip install --upgrade openpyxl
%pip install --upgrade pandas
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
# Step 2: load and clean the Data
data = pd.read_excel('final_merged_with_NULLs.xlsx')
# Replace "NULL" with NaN to ensure compatibility with numerical processing
data.replace("NULL", np.nan, inplace=True)
```

```python
# Step 3: feature selection
# Select features relevant for clustering (e.g., socio-economic indicators)
features = [
    'Expected Years of Schooling (years)', 'Mean Years of Schooling (years)', 
    'Gross National Income Per Capita (2017 PPP$)', 'Life Expectancy at Birth (years)'
]
X = data[features].fillna(data[features].median())  # Fill missing values
```

step 3 is to choose indicators that may represent the socio-economic profile of each country

```python
# Step 4: standardize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

below are techniques for evaluating the optimal number of clusters (k) in clustering algorithms, especially in KMeans clustering.

Elbow Method helps determine the ideal number of clusters by examining the Sum of Squared Errors (SSE) for each k value
    Interpretion of the Plot: The "elbow" is the optimal number of clusters, as it represents a balance between minimizing SSE and keeping k as low as possible.

Silhouette Score measures how well-separated clusters are and is another method for assessing the quality of clustering
- Generally, silhouette scores above 0.5 are considered good, while scores below 0.3 indicate poor separation
    Interpretion of the Plot: The optimal k will have the highest average silhouette score, indicating that the data points are well-clustered and clusters are distinct

---------------------------------------------------------------------------------------------------------------------------------------------
Choosing k: Look at both the Elbow and Silhouette plots. Ideally, the optimal k should correspond to a sharp "elbow" in the SSE plot and a high silhouette score, indicating compact and well-separated clusters.

```python
# Step 5: Determine Optimal Number of Clusters
# Use the Elbow Method and Silhouette Score for optimal cluster count
sse = []
silhouette_scores = []

# Test different numbers of clusters
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

# Plot the Elbow Method for SSE
plt.figure(figsize=(10, 5))
plt.plot(range(2, 11), sse, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
plt.title('Elbow Method for Optimal k')
plt.show()

# Plot the Silhouette Scores
plt.figure(figsize=(10, 5))
plt.plot(range(2, 11), silhouette_scores, marker='o', color='orange')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal k')
plt.show()
```

```python
# Step 6: Apply KMeans with Optimal k (e.g., k=3 based on plots)
optimal_k = 4  # Replace with the optimal number based on the plots
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
data['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# Step 7: Analyze and Interpret Clusters
# View average characteristics per cluster, using only numeric columns
numeric_features = data[features].select_dtypes(include=['float64', 'int64']).columns
cluster_summary = data[['Cluster'] + list(numeric_features)].groupby('Cluster').mean()
print(cluster_summary)

```

Step #7: averages each feature within clusters to understand the general characteristics of each group

to interpret: examine the average feature values for each cluster, identify the defining socio-economic characteristics
eg.
-  cluster with high income, high schooling years, and long life expectancy likely represents higher-developed countries
cluster with lower income and schooling years might represent countries with developing or emerging economies

```python
# to display unique countries in each cluster
for cluster_num in sorted(data['Cluster'].unique()):
    print(f"\nCountries in Cluster {cluster_num}:")
    countries_in_cluster = data[data['Cluster'] == cluster_num]['Country'].unique()
    print(countries_in_cluster)
```

```python
# add the clusters to the existing excel sheet
new_file_path = 'clustering_data.csv'
data.to_csv(new_file_path, index=False)
```

```python
from sklearn.metrics import davies_bouldin_score, calinski_harabasz_score

# Step 8: Evaluate Clustering Model with Additional Metrics
# Davies-Bouldin Index (DBI)
dbi = davies_bouldin_score(X_scaled, data['Cluster'])
print("Davies-Bouldin Index (DBI):", dbi)

# Calinski-Harabasz Index (CHI)
chi = calinski_harabasz_score(X_scaled, data['Cluster'])
print("Calinski-Harabasz Index (CHI):", chi)

# Silhouette Score (already calculated in the loop above, but added here for clarity)
sil_score = silhouette_score(X_scaled, data['Cluster'])
print("Silhouette Score:", sil_score)
```
